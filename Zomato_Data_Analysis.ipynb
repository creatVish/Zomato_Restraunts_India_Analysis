```python
plt.boxplot(df['average_cost_for_two'], vert=False)
plt.show()
```

In the previous plot we can see clearly there are outliers and we are going to remove outliers by IQR method

```python
q1 = df['average_cost_for_two'].quantile(0.25)
```

```python
q3 = df['average_cost_for_two'].quantile(0.75)
```

```python
IQR = q3 - q1
```

```python
IQR
```

```python
upper_limit = q3 + 1.5 * IQR
upper_limit
```

```python
lower_limit = q1 - 1.5 * IQR
lower_limit
```

```python
cleaned_data = df.loc[(df['average_cost_for_two'] >= lower_limit) & (df['average_cost_for_two'] <= upper_limit)]
```

```python
plt.boxplot(cleaned_data['average_cost_for_two'], vert=False)

plt.show()
```

Now data distribution is normal
